```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("test_image.png")
cv2_imshow(image)
```

```python
plt.imshow(image)
plt.show()
```

```python
image.shape
```

```python
image_resize = cv2.resize(image,(500,500),interpolation = cv2.INTER_NEAREST)
plt.imshow(image_resize)
plt.show()
```

```python
image_resize.shape
```

```python
image_rgb = cv2.cvtColor(image_resize, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
image_gray = cv2.cvtColor(image_resize, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
plt.show()
```

```python
cv2.imwrite("Resized_Image.jpeg",image_resize) # Saving the image
```

```python
rot, thresh1 = cv2.threshold(image_gray,120,255,cv2.THRESH_BINARY)
plt.imshow(thresh1)
plt.show()
```

```python
rot, thresh2 = cv2.threshold(image_gray,120,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh2)
plt.show()
```

```python
B,G,R = cv2.split(image_resize)
plt.figure(figsize=(20,60)) 
plt.subplot(1, 3, 1)
plt.imshow(B)
plt.subplot(1, 3, 2)
plt.imshow(G)
plt.subplot(1, 3, 3)
plt.imshow(R)
plt.show()
```

```python
merged_image = cv2.merge([B,G,R])
plt.imshow(merged_image)
plt.show()
```

```python
image_hsv = cv2.cvtColor(image_resize, cv2.COLOR_BGR2HSV)
plt.imshow(image_hsv)
plt.show()
```

```python
image_lab = cv2.cvtColor(image_resize, cv2.COLOR_BGR2LAB)
plt.imshow(image_lab)
plt.show()
```

```python
image_blur = cv2.blur(image_resize, (15,15))
plt.imshow(image_blur)
plt.show()
```

```python
import numpy as np

rectangle = np.zeros([300,300], dtype = "uint8")
image_rectangle = cv2.rectangle(rectangle, (25,25), (275,275), 255, -1)
plt.imshow(image_rectangle)
plt.show()
```

```python
circle = np.zeros([300,300], dtype = "uint8")
image_circle = cv2.circle(circle, (150,150), 100, 255, -1)
plt.imshow(image_circle)
plt.show()
```

```python
mask = np.zeros(image_resize.shape[:2], dtype = "uint8")
(cx,cy) = (image_resize.shape[1] // 2, image_resize.shape[0] // 2)
cv2.rectangle(mask, (cx-150, cy-150), (cx+150, cy+150), 255, -1)
masked = cv2.bitwise_and(image_resize,image_resize, mask = mask)
plt.imshow(masked)
plt.show()
```

```python
mask = np.zeros(image_resize.shape[:2], dtype = "uint8")
(cx,cy) = (image_resize.shape[1] // 2, image_resize.shape[0] // 2)
cv2.circle(mask, (cx, cy), 200, 255, -1)
masked = cv2.bitwise_and(image_resize,image_resize, mask = mask)
plt.imshow(masked)
plt.show()
```
